## Ablation Figures (run after notebook 04)

```python
from src.analysis.plot_results import plot_ablation_alpha, plot_ablation_position, plot_ablation_pca

# Load ablation results if available
import json, os
ablation_path = "../outputs/tables/ablation_position.json"
pca_path      = "../outputs/tables/ablation_pca.json"
alpha_path    = "../outputs/tables/ablation_alpha.json"

if os.path.exists(ablation_path):
    with open(ablation_path) as f:
        abl_pos = json.load(f)
    plot_ablation_position(abl_pos)
else:
    print("ablation_position.json not found, skipping. Run notebook 04 first.")

if os.path.exists(pca_path):
    with open(pca_path) as f:
        abl_pca = json.load(f)
    abl_pca_int = {int(k): v for k, v in abl_pca.items()}
    plot_ablation_pca(abl_pca_int)
else:
    print("ablation_pca.json not found, skipping. Run notebook 04 first.")

if os.path.exists(alpha_path):
    with open(alpha_path) as f:
        abl_alpha = json.load(f)
    plot_ablation_alpha(abl_alpha)
else:
    print("ablation_alpha.json not found, skipping. Run notebook 04 first.")
```
